```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import ExtraTreesClassifier
%matplotlib inline
```

```python
# read data
data = pd.read_csv('Data.csv')

# split data
X = data.iloc[:,0:20]
y = data.iloc[:,-1]
```

```python
# create model 
model = ExtraTreesClassifier()

# fit data to model
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
# prepair report data 
feature_importance = pd.Series(model.feature_importances_,index=X.columns)
```

```python
plt.ylabel('Feature name')
plt.xlabel('Score')
feature_importance.plot(kind='bar',colormap='plasma',figsize=(15,10))
plt.savefig('Feature_importaince.png')
```
